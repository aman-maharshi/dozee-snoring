```python
diff1 = []
for x in range(500, 598):
    diff1.append(data1['vibrations'][x+1] - data1['vibrations'][x])
```

```python
# max and min differnece for non snoring part
print max(diff1); print min(diff1)
```

```python
# max and min differnece for snoring part
print max(diff2); print min(diff2)
```

From the above two code snippets we can see that when we compute difference to two consecutive points, the result is different for the snoring part and non-snoring part of graph
For non-snoring part - maximum differnce is 3
For snoring part - maximum difference is 6

```python
array containing
diff = []
for x in range(0, 10000):
    diff.append(data['vibrations'][x+1] - data['vibrations'][x])
```

above code snippet returns a list containing the differences of two consecutive points

```python
# finding the index of the point at which first three snoring starts
li = []
for x in range(len(diff)-1):
    if diff[x] > 3:
        li.append(x)
        break   
for x in range(x+500, len(diff)-1):
    if diff[x] > 3:
        li.append(x)
        break
for x in range(x+500, len(diff)-1):
    if diff[x] > 3:
        li.append(x)
        break
for x in range(x+500, len(diff)-1):
    if diff[x] > 3:
        li.append(x)
        break


print li
```

the above code snippet returns the list of point at which snoring starts

```python
plt.figure(figsize=(20, 5))
plt.plot(data[0:5000])
for x in li:
    plt.plot(data[x: x+500], color='red')

#plt.plot(data[618: 618+500],color='red')
#plt.plot(data[1698: 1698+500],color='red')
#plt.plot(data[2775: 2775+500],color='red')
```
